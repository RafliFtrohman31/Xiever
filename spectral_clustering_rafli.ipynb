```python
from sklearn.datasets import make_moons
from sklearn.cluster import SpectralClustering
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt

# Buat dataset make_moons
X, y = make_moons(n_samples=200, noise=0.07, random_state=42)
```

```python
# Terapkan Spectral Clustering
sc = SpectralClustering(n_clusters=2, affinity='nearest_neighbors', assign_labels='kmeans', random_state=42)
labels = sc.fit_predict(X)

# Visualisasi hasil clustering
plt.figure(figsize=(6, 4))
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', s=40)
plt.title("Hasil Spectral Clustering pada Dataset make_moons")
plt.xlabel("Fitur 1")
plt.ylabel("Fitur 2")
plt.grid(True)
plt.show()
-
```

```python
# Evaluasi menggunakan Silhouette Score
score = silhouette_score(X, labels)
print(f"Silhouette Score: {score:.2f}")
```
